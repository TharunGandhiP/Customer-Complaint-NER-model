```python
import matplotlib.pyplot as plt
import numpy as np

# Data to plot
labels = ['Pre-trained NER', 'Custom-built NER']
precision = [31.00, 93.82]
recall = [29.42, 93.52]
f1_score = [57.73, 93.64]

# Create figure and axes
fig, ax = plt.subplots()

# Set the bar width
bar_width = 0.16

# Set the opacity
opacity = 0.8

# Set the positions of the bars
index = np.arange(len(labels))

# Plotting the bars
rects1 = plt.bar(index, precision, bar_width, alpha=opacity, color='b', label='Precision')
rects2 = plt.bar(index + bar_width, recall, bar_width, alpha=opacity, color='g', label='Recall')
rects3 = plt.bar(index + 2 * bar_width, f1_score, bar_width, alpha=opacity, color='r', label='F1 Score')

# Adding labels and title
plt.xlabel('SpaCy Model')
plt.ylabel('Scores')
plt.title('Results of pre-trained SpaCy vs Custom-built model')
plt.xticks(index + bar_width, labels)
plt.legend()

# Adding the data labels on top of the bars
def add_labels(rects):
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width()/2., 1*height, '%0.2f' % float(height), ha='center', va='bottom')

add_labels(rects1)
add_labels(rects2)
add_labels(rects3)

# Display the plot
plt.tight_layout()
plt.show()
```
